```python
YEARS = np.arange(2020,2027,1)
FUNDINGS =(500, 300, 200, 500, 400, 300, 200)
PERS =(150, 150, 100, 150, 150, 100, 100)
OPER =(50, 50, 50, 100, 100, 100, 100)
```

```python
def set_fonts(ax, fontsize=20):
    for item in ([ax.title, ax.xaxis.label, ax.yaxis.label] +
             ax.get_xticklabels() + ax.get_yticklabels()):
        item.set_fontsize(fontsize)
```

```python
fig = plt.figure(figsize=(10,10))
ax  = fig.add_subplot(1, 1, 1)
set_fonts(ax, fontsize=30)

plt.errorbar(YEARS,FUNDINGS, fmt="kp", ms=15, ls='none') 
plt.errorbar(YEARS,PERS, fmt="rp", ms=15, ls='none') 
plt.errorbar(YEARS,OPER, fmt="bp", ms=15, ls='none') 
plt.xlabel('year')
plt.ylabel('Costs (keuro)')
plt.show()
```

```python
fig = plt.figure(figsize=(10,10))
ax  = fig.add_subplot(1, 1, 1)
set_fonts(ax, fontsize=30)

N=7
CONS =np.array((300, 175, 50, 400, 300, 100, 100))
PERS =np.array((150, 150, 150, 150, 150, 150, 150))
OPER =np.array((50, 50, 50, 100, 100, 100, 100))

ind = np.arange(N)    # the x locations for the groups
width = 0.35       # the width of the bars: can also be len(x) sequence

p1 = plt.bar(ind, PERS, width)
p2 = plt.bar(ind, OPER, width, bottom=PERS)
p3 = plt.bar(ind, CONS, width, bottom=PERS+OPER)

plt.ylabel('Funding per year')
plt.title('Years')
plt.xticks(ind, ('2020', '2021', '2022', '2023', '2024', '2025','2026'))
plt.yticks(np.arange(0, 700, 100))
plt.legend((p1[0], p2[0], p3[0]), ('Personnel', 'Operation','Construction'))
```
